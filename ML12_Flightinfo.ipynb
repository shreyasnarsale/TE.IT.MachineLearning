```python
# Drop columns that are almost completely null
data = data.drop(columns=['Unnamed: 22', 'NA', 'NA.1', 'NA.2', 'NA.3', 'NA.4'])
```

```python
data.columns
```

```python
data.columns = ['FlightDate', 'AirlineCode', 'FlightNumber', 'TailNumber', 'OriginAirportCode',
              'DestAirportCode', 'ScheduledDepTime', 'DepTime', 'DepDelay', 'TaxiOut',
              'WheelsOff', 'WheelsOn', 'TaxiIn', 'ScheduledArrTime', 'ArrTime', 'ArrDelay',
              'Cancelled', 'CancellationCode', 'Diverted', 'AirTime', 'Distance',
              'CarrierDelay', 'WeatherDelay']
```

```python
data.columns
```

```python
# Replace nulls with mean safely
numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns

for col in numeric_cols:
    data[col] = data[col].fillna(data[col].mean())

#  fill non-numeric columns with 'Unknown'
non_numeric_cols = data.select_dtypes(exclude=['float64', 'int64']).columns
data[non_numeric_cols] = data[non_numeric_cols].fillna('Unknown')
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
# Select numeric columns for clustering
numeric_cols = data.select_dtypes(include=['int64', 'float64']).columns
X = data[numeric_cols]

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# Add cluster labels to your DataFrame
data['Cluster'] = clusters
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=data['Cluster'], palette='Set1')
plt.title("K-Means Clustering (5 Clusters)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.show()
```

```python
# Select numeric columns
numeric_cols = data.select_dtypes(include=['int64', 'float64']).columns
X = data[numeric_cols]

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-Means
kmeans = KMeans(n_clusters=5, random_state=42)
data['Cluster'] = kmeans.fit_predict(X_scaled)

# PCA for 2D visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Transform centroids to PCA space
centroids_pca = pca.transform(kmeans.cluster_centers_)

# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=data['Cluster'], palette='Set1', s=50)

# Plot centroids
plt.scatter(centroids_pca[:,0], centroids_pca[:,1], s=200, c='black', marker='*', label='Centroids')

plt.title("K-Means Clustering with PCA (5 Clusters)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend()
plt.show()
```
