```python
# Select numeric columns for clustering
x = df[['age', 'balance', 'duration']].values

# Initialize and fit K-Means
km = KMeans(
    n_clusters=3,          # you can change this to 4 or 5 if you want more clusters
    init='k-means++',
    max_iter=300,
    n_init=10,
    random_state=0
)

y_means = km.fit_predict(x)

# Plot clusters
plt.figure(figsize=(10,6))
plt.scatter(x[y_means == 0, 0], x[y_means == 0, 1], s=100, c='pink', label='Cluster 1')
plt.scatter(x[y_means == 1, 0], x[y_means == 1, 1], s=100, c='yellow', label='Cluster 2')
plt.scatter(x[y_means == 2, 0], x[y_means == 2, 1], s=100, c='cyan', label='Cluster 3')

# Plot centroids
plt.scatter(
    km.cluster_centers_[:, 0],
    km.cluster_centers_[:, 1],
    s=200,
    marker='*',
    c='red',
    label='Centroids'
)

plt.title('Customer Clusters (Bank Dataset)')
plt.xlabel('Age')
plt.ylabel('Balance')
plt.legend()
plt.show()

# Output centers, labels, and inertia
{
    "Cluster Centers": pd.DataFrame(km.cluster_centers_, columns=['age', 'balance', 'duration']),
    "Cluster Labels": y_means,
    "Inertia": km.inertia_
}
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch

# Example numeric features from your dataset
# Make sure x is defined as numeric data
# x = df[['age','balance','duration','campaign','pdays','previous']].values

plt.figure(figsize=(12, 6))
plt.title("Truncated Hierarchical Clustering Dendrogram")
plt.xlabel("Cluster Size")
plt.ylabel("Euclidean Distance")

# Create linkage matrix
linkage_matrix = sch.linkage(x, method='ward')

# Plot a truncated dendrogram (only last few merges)
sch.dendrogram(
    linkage_matrix,
    truncate_mode='lastp',   # show only the last p merged clusters
    p=12,                    # number of clusters to show (adjust as needed)
    leaf_rotation=90.,
    leaf_font_size=10.,
    show_contracted=True     # shows the contracted branches
)

plt.show()
```
