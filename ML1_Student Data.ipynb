```python
# Take only first 16 columns
subset_df = df.iloc[:, :16]

# Remove duplicate column names
subset_df = subset_df.loc[:, ~subset_df.columns.duplicated()]

# Keep only numeric columns
numeric_subset = subset_df.select_dtypes(include=['number']).copy()

# Drop columns with arrays instead of flat numbers
for col in numeric_subset.columns:
    if numeric_subset[col].apply(lambda x: isinstance(x, (list, tuple, np.ndarray))).any():
        numeric_subset.drop(columns=[col], inplace=True)


sns.pairplot(numeric_subset)
plt.show()
```
